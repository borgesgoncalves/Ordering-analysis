#### Frequencia de status

```python
type_status_count = orders_analytics['order_status'].value_counts()
total_orders = type_status_count.sum()
finished_orders_count = type_status_count.get('FINISHED', 0)
canceled_orders_count = type_status_count.get('CANCELED', 0)
normalized_finished = finished_orders_count / total_orders
normalizes_canceled = canceled_orders_count / total_orders

print(f"Total de Pedidos: {total_orders}")
print(f"Pedidos Finalizados: {finished_orders_count} ({normalized_finished:.2%})")
print(f"Pedidos Cancelados: {canceled_orders_count} ({normalizes_canceled:.2%})")

plt.figure(figsize=(8, 6))
type_status_count.plot(kind='bar', color='skyblue', label=f'finished: {normalized_finished:.2%}, canceled: {normalizes_canceled:.2%}')
plt.title('Contagem de Status dos Pedidos')
plt.xlabel('Status')
plt.ylabel('Contagem')
plt.xticks(rotation=360)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
# Quantas lojas tem em cada hub?
df_store_count = df_stores_hubs.groupby('hub_id').agg(
    {'store_id': 'count',
     'store_segment': lambda x: (x == 'FOOD').sum(),
     'hub_name': 'first',
     'hub_city': 'first',
     }).reset_index()

df_store_count.rename(columns={'store_id': 'store_count'}, inplace=True)
df_store_count.rename(columns={'store_segment': 'food_segment'}, inplace=True)
df_store_count['good_segment'] = df_stores_hubs.groupby('hub_id')['store_segment'].apply(lambda x: (x == 'GOOD').sum()).values

df_store_count.sort_values(by='store_count', ascending=False, inplace=True)
df_store_count.reset_index(drop=True, inplace=True)
df_store_count
```
